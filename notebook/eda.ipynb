```python
df = pd.read_excel(excel, sheet_name='input_vacation')
```

```python
RPS = ['RP1', 'RP2', 'RP3', 'RP4', 'RP5', 'RP6', 'RP7', 'RP8', 'RP9', 'RP10', 'RP11', 'RP12']
RRS = ['RR1', 'RR2', 'RR3', 'RR4', 'RR5', 'RR6', 'RR7', 'RR8', 'RR9', 'RR10', 'RR11', 'RR12']

def f(row, PR):
    if row[PR] == 0:
        return 0
    return (4 - row[RP]) * (70 + 1) + row['Pers Level']  # int


for RP in RPS:
    df[f'{RP}_score'] = df.apply(f, axis=1, args=(RP,))
```

```python
from collections import defaultdict

q = defaultdict()

for qual in quals:
    q[qual] = df.loc[df[qual] == 1]
```

```python
work_df = pd.DataFrame(index=req_work.index, columns=range(1,13))
```

```python
def f(row, RP, RR):
    if row[RP] == 0:
        row[f'Work time {RP}'] = row['Max Fly']
        row['Vacation time'] += 0
    else:
        row['Vacation time'] += row[RR]
        row[f'Work time {RP}'] = row['Max Fly'] - row[RR]
    return row


for qual in quals:
    
    for i, (RP, RR) in enumerate(zip(RPS, RRS), 1):
        _df[f'Work time {RP}'] = _df['Vacation time'] = 0
        q[qual] = q[qual].apply(f, args=(RP, RR), axis=1)
        work_df.loc[qual, i] = q[qual][f'Work time {RP}'].sum()
```
